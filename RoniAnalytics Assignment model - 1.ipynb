```python
model.predict(data[['amount']])
model_pred_price = model.predict(data[['amount']])
actual_price = data['price']
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
mean_squared_error(actual_price, model_pred_price)
```

```python
mean_absolute_error(actual_price, model_pred_price) #MAE
```

```python
np.sqrt(mean_squared_error(actual_price, model_pred_price))
```

```python
r2_score(actual_price, model_pred_price)
```

```python
def mean_absolute_percentage_error(y_true, y_pred): 
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

mean_absolute_percentage_error(actual_price, model_pred_price)
```
